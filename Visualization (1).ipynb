### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns
np.random.seed(42)
num_samples = 100
num_features = 5
X = np.random.randn(num_samples, num_features)
corr_matrix = np.array([[1, 0.8, 0.5, 0.3, 0.1],
                        [0.8, 1, 0.7, 0.4, 0.2],
                        [0.5, 0.7, 1, 0.6, 0.3],
                        [0.3, 0.4, 0.6, 1, 0.5],
                        [0.1, 0.2, 0.3, 0.5, 1]])
X_correlated = X @ np.linalg.cholesky(corr_matrix).T
feature_names = [f'Feature {i+1}' for i in range(num_features)]
df = pd.DataFrame(X_correlated, columns=feature_names)
corr_df = df.corr()
sns.heatmap(corr_df, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns

```python
output_notebook()
np.random.seed(42)  # For reproducibility
num_points = 100
x = np.random.uniform(0, 10, num_points)
y = np.random.uniform(0, 10, num_points)
sizes = np.random.uniform(5, 50, num_points)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], num_points)  # Bubble colors
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')
p.scatter(x='x', y='y', size='sizes', color='colors', legend_field='colors', fill_alpha=0.6, source=df)
p.legend.title = 'Colors'
p.legend.label_text_font_size = '10pt'
p.legend.location = 'top_left'
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

p.vbar(x='fruits', top='counts', width=0.6, source=source, color='navy', alpha=0.7)

p.grid.grid_line_color = None
p.xaxis.major_label_orientation = "vertical"
p.y_range.start = 0

show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

np.random.seed(42)  # For reproducibility
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.7)

p.grid.grid_line_color = None
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12

show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np
import pandas as pd

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

df = pd.DataFrame(data)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')

mapper = linear_cmap(field_name='value', palette='Viridis256', low=min(data['value']), high=max(data['value']))

p.rect(x='x', y='y', width=0.1, height=0.1, source=df, fill_color=mapper, line_color=None)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

show(p)
```
